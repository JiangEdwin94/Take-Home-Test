```python
import pandas as pd
```

```python
loyalty = pd.read_csv('loyalty.csv')
loyalty = loyalty.add_prefix('loyalty_')
transaction = pd.read_csv('transactions.csv')

df = pd.merge(transaction,loyalty, left_on='id', right_on='loyalty_id')
df = df[['name', 'city', 'phone-number', 'email',
       'loyalty_license-plate','Amount']]
df.head(5)
```

```python
#find the spending power for each city
City_amount = df.groupby(['city'])['Amount'].agg('sum')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline 
```

```python
x = df[['Amount']]
```

```python
#finding optimal number of clusters using the elbow method  
from sklearn.cluster import KMeans  
wcss_list= [] 
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  
```

```python
#training the K-means model 
kmeans = KMeans(n_clusters=2, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x)  
```

```python
y_predict = pd.DataFrame(y_predict)
df2 = pd.concat([df, y_predict], axis=1)
df2.rename(columns = {0:'Cluster'}, 
            inplace = True)
df2
```
